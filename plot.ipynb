```python
import pandas as pd
import os
import matplotlib.pyplot as plt


def plot_and_stat(key, plot_idx):
    result = []
    plt.subplot(plot_idx)
    for item in os.listdir(exp_dir):
        if item.startswith(pre) and item.endswith(pre2):
            df = pd.read_csv(f'{exp_dir}/{item}/log.txt', delimiter='\t')
            plt.plot(df[key].ravel(),label=item)
            result.append((item, df[key].min()))
    plt.title(key)
    plt.legend()
    return sorted(result, key=lambda x:x[-1])
```

```python
pre = ''
pre2 = ''
exp_dir = 'experiments/h36m_sh/length'
plt.figure(figsize=(16, 16))

mpose = plot_and_stat('Pose MPJPE', 221)
ppose = plot_and_stat('Pose P-MPJPE', 222)
mmotion = plot_and_stat('Motion MPJPE', 223)
pmotion = plot_and_stat('Motion P-MPJPE', 224)
plt.show()
print('pose', mpose)
print()
print('p-pose', ppose)
print()
print('motion', mmotion)
print()
print('p-motion', pmotion)
print()
```
